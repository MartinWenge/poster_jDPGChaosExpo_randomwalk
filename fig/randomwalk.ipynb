```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def randomWalk(nSteps=10):
    walkX, walkY = [0],[0]
    for _ in range(nSteps):
        x,y = np.random.randint(-1,2,2)
        walkX.append(walkX[-1]+x)
        walkY.append(walkY[-1]+y)
        
    return walkX, walkY
```

```python
plt.rcParams["font.family"] = "serif"
plt.figure(dpi=600)
x,y = randomWalk(2560)
plt.plot(x,y, 'm-o', ms=0.1,linewidth=0.5, label="2560")
x,y = randomWalk(640)
plt.plot(x,y, 'b-o', ms=0.2,linewidth=0.8, label="640")
x,y = randomWalk(200)
plt.plot(x,y, 'g-o', ms=0.4, linewidth=1.0, label="160")
x,y = randomWalk(40)
plt.plot(x,y, 'r-o', ms=0.6, linewidth=1.3, label="40")
plt.xticks([])
plt.yticks([])
plt.xlabel('West - Ost')
plt.ylabel('Süd - Nord')
plt.legend()
#plt.savefig('fig/randomWalkPy.png')
plt.show()
```
